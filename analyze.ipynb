```python
import pandas as pd 
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt 
```

```python
data = pd.read_csv('celldata.csv') 
```

```python
def mean_std_calc(data,num):
    mean_list = []
    std_list = []
    max_list = []
    min_list = []
    for i in range(num):
         mean = data[data.columns[i]].mean()
         mean_list.append(mean)
         std = data[data.columns[i]].std()
         std_list.append(std)
         maxcell = max(data[data.columns[i]])
         max_list.append(maxcell)
         mincell = min(data[data.columns[i]])
         min_list.append(mincell)   
    return mean_list,std_list,max_list,min_list
```

```python
a= mean_std_calc(data,len(data.columns))
```

```python
a
```

```python
def normfun(x,mu,sigma):
    pdf = np.exp(-((x - mu)**2)/(2*sigma**2)) / (sigma * np.sqrt(2*np.pi))
    return pdf
```

```python
def subplt(sub_num,data):
    for i in range(sub_num):
            x = np.arange(data[3][i],data[2][i],0.1) 
            y = normfun(x, data[0][i], data[1][i])
            plt.subplot(4,12,i+1)
            plt.plot(x,y,'r--')    
```
